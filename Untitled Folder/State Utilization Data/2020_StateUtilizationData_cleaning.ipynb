```python
import pandas as pd
import numpy as np
```

```python
# import CSV file
df = pd.read_csv(r"C:\ERICS_STUFF\StateDrugUtilization\2020_StateDrugUtilizationData.csv") 

# create list of columns to drop
to_drop = ['utilization_type', 
          'supression_used',
          'medicaid_amount_reimbursed',
          'non_medicaid_amount_reimbursed']


# drop the defined columns. 
# inplace true stating changes should be made directly
df.drop(columns=to_drop, inplace=True)

# change package_size to integer
df.fillna({'package_size':-1}, inplace=True)
df.package_size= df.package_size.astype(int)

# Get rid of all XX states
df = df[df.state != 'XX']

# drop rows where null in a specific column
df_cleansed = df.dropna(axis=0, how="all", subset=['total_amount_reimbursed'])




# display(df)
# display(df_cleansed.dtypes) - To see data types
display(df_cleansed)
df_cleansed_csv = df_cleansed.to_csv('cleansed', index=False)
```
